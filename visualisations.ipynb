```python
import matplotlib.pyplot as plt

# Model names and accuracies
models = [
    'Sannino et al (TensorFlow)',
    'Singh et al (RNN)',
    'Hannun et al (CNN)',
    'Bassiouni et al (KNN)',
    'Anwar et al (SVM)',
    'Kumar et al (MLP)'
]

accuracies = [99.68, 85.4, 98.5, 99, 99.6, 98.72]

# Creating the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(models, accuracies, color=['skyblue', 'skyblue', 'skyblue', 'lightgreen', 'lightgreen', 'lightgreen'])

# Adding title and labels
plt.title('Accuracy Comparison of Classification Models')
plt.xlabel('Model')
plt.ylabel('Accuracy (%)')

# Adding value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.2, f'{yval:.2f}%', ha='center', va='bottom')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
