```python
import os
import scipy.io as sio
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
```

```python
#getting all the hea files

ecg_data = os.path.join(os.getcwd(), '/data/processed_ecg_signals_2/WFDBRecords')

hea_files = []

for root, dirs, files in os.walk(ecg_data):
    for file in files:
        if file.endswith('.hea'):
            hea_files.append(os.path.join(root, file))
```
